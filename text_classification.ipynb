```python
import pandas as pd

books = pd.read_csv('books_cleaned.csv')
```

```python
books["categories"].value_counts().reset_index()
```

```python
# checking out the categories that have 50 books or more
books["categories"].value_counts().reset_index().query("count > 50")
```

```python
# creating a mapping of the top 12 categories to either fiction or non-fiction
category_mapping = {
    "Fiction": "Fiction",
    "Juvenile Fiction": "Children's Fiction",
    "Biography and Autobiography": "Nonfiction",
    "History": "Nonfiction",
    "Literary Criticism": "Nonfiction",
    "Philosophy": "Nonfiction",
    "Religion": "Nonfiction",
    "Comics and Graphic Novels": "Fiction",
    "Drama": "Fiction",
    "Juvenile Nonfiction": "Children's Nonfiction",
    "Science": "Nonfiction",
    "Poetry": "Fiction"
}
```

```python
books["simple_categories"] = books["categories"].map(category_mapping)
```

```python
books
```

```python
books[~(books["simple_categories"].isna())]
```

```python
from transformers import pipeline

fiction_cats = ["Fiction", "Nonfiction"]
pipe = pipeline("zero-shot-classification",
                model="facebook/bart-large-mnli",
                device=0)
```

```python
#first of the known fiction examples
sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[0]
```

```python
pipe(sequence, fiction_cats)
```

```python
import numpy as np

# get the predicted label
max_index = np.argmax(pipe(sequence, fiction_cats)["scores"])
max_label = pipe(sequence, fiction_cats)["labels"][max_index]
max_label
```

```python
def generate_predictions(sequence, categories):
    predictions = pipe(sequence, categories)
    max_index = np.argmax(predictions['scores'])
    max_label = predictions["labels"][max_index]
    return max_label
```

```python
from tqdm import tqdm

actual_cats = []
predicted_cats = []

for i in tqdm(range(0, 300)):
    sequence = books.loc[books["simple_categories"] == "Fiction", "description"].reset_index(drop=True)[i]
    predicted_cats += [generate_predictions(sequence, fiction_cats)]
    actual_cats += ["Fiction"]
```

```python
for i in tqdm(range(0, 300)):
    sequence = books.loc[books["simple_categories"] == "Nonfiction", "description"].reset_index(drop=True)[i]
    predicted_cats += [generate_predictions(sequence, fiction_cats)]
    actual_cats += ["Nonfiction"]
```

```python
predictions_df = pd.DataFrame({"Actual Categories": actual_cats, "Predicted Categories": predicted_cats})
```

```python
predictions_df.head()
```

```python
# adding a column which tells us if the predictions are correct or not
predictions_df["correct_predictions"] = (
    np.where(predictions_df["Actual Categories"] == predictions_df["Predicted Categories"], 1, 0)
)
```

```python
# check what percentage of the prdictions are correct
predictions_df["correct_predictions"].sum() / len(predictions_df)
```

```python
isbn_list = []
predicted_cats = []

missing_cats = books.loc[books['simple_categories'].isna(), ['isbn13', 'description']].reset_index(drop=True)
```

```python
missing_cats.head()
```

```python
for i in tqdm(range(0, len(missing_cats))):
    sequence = missing_cats['description'][i]
    predicted_cats += [generate_predictions(sequence, fiction_cats)]
    isbn_list += [missing_cats['isbn13'][i]]
```

```python
missing_predictions_df = pd.DataFrame({"isbn13": isbn_list, "predicted_categories": predicted_cats})
missing_predictions_df
```

```python
# merging the two prediction dataframes
books = pd.merge(books, missing_predictions_df, how="left", on="isbn13")
books["simple_categories"] = np.where(books["simple_categories"].isna(), books["predicted_categories"], books["simple_categories"])
books.drop(columns = "predicted_categories")
```

```python
# further breaking down the functional categories
books[books["categories"].str.lower().isin([
    "romance",
    "science fiction",
    "scifi",
    "fantasy",
    "horror",
    "mystery",
    "thriller"
    "comedy",
    "action",
    "crime",
    "historical"
])]
```

```python
# saving the new dataframe for further usage
books.to_csv("books_with_categories.csv", index=False)
```
