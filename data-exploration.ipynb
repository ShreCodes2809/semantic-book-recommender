```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd

books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# visualizing the missing values in each of the columns
ax = plt.axes()
sns.heatmap(books.isna(). transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing values")

plt.show()
```

```python
# check the missing descriptions and the age of the books
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2025 - books["published_year"]
```

```python
# determining if any of the columns have any correlation or nots
columns_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[columns_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize=(12, 8))
heatmap = sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", cbar_kws={"label": "Spearman Correlation"})
heatmap.set_title("Correlation Heatmap")
plt.show()
```

```python
#checking what are the exact number of columns with missing values apart from subtitle
book_missing = books[
    (~books["description"].isna()) &
    (~books["num_pages"].isna()) &
    (~books["average_rating"].isna()) &
    (~books["published_year"].isna())
]

book_missing
```

```python
# checking which categories are mostly used ones
book_missing["categories"].value_counts().reset_index().sort_values(by="count", ascending=False)
```

```python
#checking the number of words in a description
book_missing["words_in_desc"] = book_missing["description"].str.split().str.len()
```

```python
book_missing
```

```python
#checking the useless descriptions
book_missing.loc[book_missing["words_in_desc"].between(1, 4), "description"]
```

```python
book_missing.loc[book_missing["words_in_desc"].between(5, 14), "description"]
```

```python
book_missing.loc[book_missing["words_in_desc"].between(15, 24), "description"]
```

```python
book_missing_25words = book_missing[book_missing["words_in_desc"] >= 25]
```

```python
book_missing_25words
```

```python
book_missing_25words["title_and_subtitle"] = (
    np.where(book_missing_25words["subtitle"].isna(), book_missing_25words["title"],
             book_missing_25words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1))
)
```

```python
book_missing_25words
```

```python
# create a tagged description column
book_missing_25words["tagged_desc"] = book_missing_25words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
book_missing_25words
```

```python
# dropping the columns we dont need
(
    book_missing_25words
    .drop(["subtitle", "missing_description", "age_of_book", "words_in_desc"], axis=1)
    .to_csv("books_cleaned.csv", index=False)
)
```
